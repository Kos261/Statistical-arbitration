Lab I.102 Rodzina kwadratowa i wykładnik Lyapunowa
Określmy rodzinę funkcji logistycznych $f_{r}:[0,1] \to [0,1]$, jako $f_r(x) = rx(1-x)$.


A. Dla parametrów $r\in [2.5,4]$ narysuj $10^2$ ostatnich spośród $10^4$ iteracji punktu krytycznego $f_r$ w przestrzeni fazowej $[2.5,4] \times I$.

B. Określmy wykładnik Lyapunova dla punktu $x_0$ wzorem:
   $\lambda_r(x_0)=\lim_{n\to\infty}\frac{1}{n}\sum_{j=0}^{n-1}\ln|f'_{r}(f_{r}^{j}(x_0))|$
Dla parametrów $r\in [2.5,4]$ oblicz przybliżenia wykładników Lyapunova punktu krytycznego
i narysuj je w przestrzeni fazowej $[2.5,4] \times [-1,1]$.
