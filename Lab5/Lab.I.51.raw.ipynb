```python
wzrost = np.array([150, 155,  160, 165,  170, 175, 180, 150, 155, 160, 165, 170, 175, 180], dtype = float)
waga = np.array([46, 48,  51, 54,  58, 60, 64, 49, 51, 54, 57, 61, 63, 67], dtype = float)
wiek = np.array([20, 20,  20, 20,  20, 20, 20, 30, 30, 30, 30, 30, 30, 30], dtype = float)


p = round(wzrost.shape[0] * 0.8)
 
#print(wiek[:p])

wzrost_train = wzrost[:p]
waga_train = waga[:p]
wiek_train = wiek[:p]

wzrost_test = wzrost[p:]
waga_test = waga[p:]
wiek_test = wiek[p:]
```

```python
#arr = np.array(wzrost, waga, dtype='float32')
array = np.vstack((wzrost,waga))
```

```python
model = tf.keras.models.Sequential([
  #tf.keras.layers.Flatten(input_shape=(28, 28)),
  tf.keras.layers.Flatten(input_shape= (2,1)),
  tf.keras.layers.Dense(3, activation='relu'),
  #tf.keras.layers.Dropout(0.2),
  tf.keras.layers.Dense(1)
])
```

```python
model.compile(optimizer='rmsprop',
loss='mse',
metrics=['mae'])
#model.compile(optimizer='rmsprop', loss='mse', metrics=['mae'])
```

```python
wiek_train.T
```

```python
#model.fit(wzrost_train, wiek_train, epochs=80, batch_size=10, verbose=1)
model.fit(array_train, wiek_train, epochs=500, batch_size=10, verbose=1)

#model.fit(train_data, train_targets,
#epochs=80, batch_size=16, verbose=0)
```

```python
model.evaluate(array_test,  wiek_test, verbose=2)
```
